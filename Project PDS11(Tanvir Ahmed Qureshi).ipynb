```python
#Answer to the Q.N-4:
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("https://raw.githubusercontent.com/PulockDas/pd-12-resources/refs/heads/master/titanic.csv")

survival_counts = df.groupby(['Survived', 'Sex']).size().unstack()

survival_counts.plot(kind='bar', stacked=True, color=['blue', 'pink'])

plt.title("Titanic Survival Count by Gender")
plt.xlabel("Survival Status (0 = Dead, 1 = Survived)")
plt.ylabel("Number of Passengers")
plt.legend(title="Gender", labels=['Male', 'Female'])
plt.xticks(ticks=[0, 1], labels=['Dead', 'Survived'], rotation=0)


plt.show()
```

```python
#Answer to the Q.N-8:
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("https://raw.githubusercontent.com/PulockDas/pd-12-resources/refs/heads/master/titanic.csv")


df['AgeClass'] = df['Age'].apply(lambda age: 0 if age <= 16 else 
                                            1 if age <= 26 else 
                                            2 if age <= 36 else 
                                            3 if age <= 62 else 4)


df.drop(columns=['Age'], inplace=True)


survival_by_age = df.groupby(['AgeClass', 'Survived', 'Sex']).size().unstack()


survival_by_age.plot(kind='bar', stacked=True, color=['blue', 'pink'], figsize=(8, 6))


plt.title("Titanic Survival by Age Class and Gender")
plt.xlabel("Age Class and Survival Status (0 = Dead, 1 = Survived)")
plt.ylabel("Number of Passengers")
plt.legend(title="Gender", labels=['Male', 'Female'])
plt.xticks(rotation=0)


plt.show()
```

```python
#Assignment:
#2 No:
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("https://raw.githubusercontent.com/PulockDas/pd-12-resources/refs/heads/master/titanic.csv")


dead_passengers = df[df['Survived'] == 0]


plt.figure(figsize=(8, 6))


plt.scatter(dead_passengers[dead_passengers['Sex'] == 'male']['Age'], 
            dead_passengers[dead_passengers['Sex'] == 'male']['Fare'], 
            color='blue', label='Male', alpha=0.6)


plt.scatter(dead_passengers[dead_passengers['Sex'] == 'female']['Age'], 
            dead_passengers[dead_passengers['Sex'] == 'female']['Fare'], 
            color='yellow', label='Female', alpha=0.6)

plt.xlabel("Age")
plt.ylabel("Fare")
plt.title("Scatter Plot of Dead Male and Female Passengers (Titanic)")
plt.legend()
plt.grid(True)


plt.show()

```

```python
#Answer to the Q N-4:
import pandas as pd


df_tt1 = pd.read_csv("term-test-1-result.csv")
df_tt2 = pd.read_csv("term-test-2-result.csv")


merged_df = pd.merge(df_tt1, df_tt2, on="Registration Number")


print("Columns in merged dataframe:", merged_df.columns)


merged_df.rename(columns={'TT-1 Marks_x': 'TT-1 Marks', 'TT-2 Marks_y': 'TT-2 Marks'}, inplace=True)

print("Updated columns in merged dataframe:", merged_df.columns)


merged_df['Best Marks'] = merged_df[['TT-1 Marks', 'TT-2 Marks']].max(axis=1)


merged_df['Average Marks'] = merged_df[['TT-1 Marks', 'TT-2 Marks']].mean(axis=1)


print(merged_df[['Registration Number', 'Name', 'TT-1 Marks', 'TT-2 Marks', 'Best Marks', 'Average Marks']].head())
```

```python
#Answer to the Q N-5:
merged_df.drop(columns=['TT-1 Marks', 'TT-2 Marks'], inplace=True)


print(merged_df[['Registration Number', 'Name', 'Best Marks', 'Average Marks']].head())
```

```python
#Answer to the Q N-6:
merged_df.to_csv('final term test result.csv', index=False)


print("The dataframe has been saved to 'final term test result.csv'")
```

```python
#Answer to the Q N-7:
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


merged_df = pd.read_csv("final term test result.csv")


X = merged_df[['Average Marks']].values  # Reshape it to fit the model


kmeans = KMeans(n_clusters=3, random_state=42)
merged_df['Cluster'] = kmeans.fit_predict(X)


plt.figure(figsize=(8, 6))
plt.scatter(merged_df['Average Marks'], [1]*len(merged_df), c=merged_df['Cluster'], cmap='viridis', s=100)
plt.title("K-Means Clustering of Average Marks")
plt.xlabel("Average Marks")
plt.yticks([])
plt.colorbar(label='Cluster')
plt.show()


print("Cluster Centers:", kmeans.cluster_centers_)
```
